# Working Github Repo API 

```python
#api endpoints
data_url = 'https://raw.githubusercontent.com/odhiambokevin/luxdev_staff_python_project/refs/heads/main/staff_data.json'
data_url2 = 'https://github.com/odhiambokevin/luxdev_staff_python_project/blob/main/staff_data.json'
```

```python
#import requests
import requests
import pandas as pd
```

## Get the data from the url using requests

```python
get_raw_data = requests.get(data_url)

if get_raw_data.status_code == 200:
    #parse json data into a dictionary or list
    raw_data = get_raw_data.json()
    print("Data retrieved successfully!")
else:
    print(f"Failed to retrieve products data. Status code: {get_raw_data.status_code}")
```

```python
#get the raw products data as is
data_df_raw = pd.DataFrame(raw_data)
print(data_df_raw.head())
```

```python
from pathlib import Path
output_dir = Path('data')
file_path = output_dir / 'github_data.csv'

#create the output folder if it doesn't exist
output_dir.mkdir(parents=True, exist_ok=True)
```

```python
#export to csv
data_df_raw.to_csv(file_path,index=False)
```
