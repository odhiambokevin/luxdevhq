# Working WithDummy JSON Data With API 

```python
#api endpoints
products_url = 'https://dummyjson.com/products'
carts_url = 'https://dummyjson.com/carts'
```

```python
#import requests
import requests
import pandas as pd
```

### Retrieve both products and cart data in clean flat tables

After retrieving data, I use .json_normalize so that nested dictionaries inside columns are extracted into own separate columns.

```python
get_products_data = requests.get(products_url)

if get_products_data.status_code == 200:
    #parse json data into a dictionary or list
    products_data = get_products_data.json()
    print("Products retrieved successfully!")
else:
    print(f"Failed to retrieve products data. Status code: {get_products_data.status_code}")
```

```python
#get the raw products data as is
products_df_raw = pd.DataFrame(products_data)
print(products_df_raw.head())

#flatten the raw data to get a clean flat table(prevent dict in columns)
products_df = pd.json_normalize(products_df_raw['products'])
products_df.head()
```

```python
get_carts_data = requests.get(carts_url)

if get_carts_data.status_code == 200:
    #parse json data into a dictionary or list
    carts_data = get_carts_data.json()
    print("Cart data retrieved successfully!")
else:
    print(f"Failed to retrieve carts data. Status code: {get_carts_data.status_code}")
```

```python
#get the raw carts data as is
carts_df_raw = pd.DataFrame(carts_data)
print(carts_df_raw.head())

#flatten the raw data to get a clean flat table(prevent dict in columns)
carts_df = pd.json_normalize(carts_df_raw['carts'])
df_exploded = carts_df.explode('products').reset_index(drop=True)
products_normalized = pd.json_normalize(df_exploded['products'])
final_df = pd.concat([df_exploded.drop(columns=['products']), products_normalized], axis=1)
final_df.head()

```

### Export data to csv

```python
from pathlib import Path
output_dir = Path('data')
product_file_path = output_dir / 'product_data.csv'
cart_file_path = output_dir / 'cart_data.csv'

#create the output folder if it doesn't exist
output_dir.mkdir(parents=True, exist_ok=True)
```

```python
#export to csv
products_df.to_csv(product_file_path,index=False)
carts_df.to_csv(cart_file_path,index=False)
```
